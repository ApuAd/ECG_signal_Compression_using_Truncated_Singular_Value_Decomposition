```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
data=pd.read_csv("/content/ptbdb_normal.csv",header=None)
```

```python
# Variance Analysis :

pca=PCA()  
pca.n_components=50
ind=[]
vare=[]
pca_data=pca.fit_transform(data)  
# look at explainded variance of PCA components 
exp_var_cumsum=pd.Series(np.round(pca.explained_variance_ratio_.cumsum(),4)*100)  
for index,var in enumerate(exp_var_cumsum):  
    print('if n_components= %d,   variance=%f' %(index,np.round(var,3)))
    ind.append(index)
    vare.append(np.round(var,3))
```

```python
pd.Series(vare).plot()
plt.title("Variance Analysis")
plt.xlabel("N_components")
plt.ylabel("Variance");
```

```python
import sklearn.decomposition

X =data.copy()

mu = np.mean(X, axis=0)

pca = sklearn.decomposition.PCA()
pca.fit(X)  # Fitting PCA on the data

nComp = 35   # Choosing n_components for the output data

X_pca =pca.transform(X)[:,:nComp]
```

```python
Xhat = np.dot(X_pca, pca.components_[:nComp,:])
Xhat += mu
```

```python
X_pca.shape
```

```python
pd.DataFrame(X_pca).head() # Compressed Data
```

```python
import sys
def sizereduction(new_df,old_df,one,two):
    print(f"Compression Analysis of {one} vs {two}\n")
    one_size=sys.getsizeof(new_df)
    two_size=sys.getsizeof(old_df)
    
    print(f"{one} data size in bytes: {one_size}")
    print(f"{one} data size in KB: {one_size/1024}KB \n")
    
    print(f"{two} data size in bytes: {two_size}")
    print(f"{two} data size in KB: {two_size/1024}KB \n")
    
    comp_ratio = (one_size/two_size)*100 
    print(f"{comp_ratio}% is the compressed ratio of the {one} dataframe")
    print(f"{100 -comp_ratio}% of space is conserved compared to the {two} dataframe ")
```

```python
X_pca_df = pd.DataFrame(X_pca)
```

```python
sizereduction(X_pca_df,X,"PCA compressed Data","uncompressed data")
```

```python
X_pca_df_c = X_pca_df.copy()
```

```python
for i in X_pca_df_c.columns:
    X_pca_df_c[i]= pd.cut(X_pca_df_c[i],bins=200)
```

```python
X_pca_df_c.head()
```

```python
for i in X_pca_df_c.columns:
    X_pca_df_c[i] = X_pca_df_c[i].apply(lambda x : float(x.mid))
```

```python
# X_pca_df_c is the Quantized Data
sizereduction(X_pca_df_c,X,"Quantized Data","Original Data")
```
